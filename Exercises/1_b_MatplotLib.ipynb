## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit


```python
#summe = sum(data.total_profit)
#print(summe)
```

```python
fig, ax = plt.subplots()
ax.plot(data.month_number, data.total_profit)

ax.set(xlabel='time (month)', ylabel='profit (€)',
       title='Total profit per month')
ax.grid()
```

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


```python
fig, ax = plt.subplots()
ax.plot(data.month_number, data.total_profit, ':ro', linewidth=3)

ax.set(xlabel='time (month)', ylabel='profit (€)',
       title='Total profit per month')
ax.legend(['Profit'], loc='lower right')
plt.xticks(data.month_number)
plt.yticks([100000, 200000, 300000, 400000, 500000])

ax.grid()
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

```python
fig, ax = plt.subplots()
ax.plot(data.month_number, data.facecream, ':r', linewidth=2)
ax.plot(data.month_number, data.facewash, ':b', linewidth=2)
ax.plot(data.month_number, data.toothpaste, ':g', linewidth=2)
ax.plot(data.month_number, data.bathingsoap, ':c', linewidth=2)
ax.plot(data.month_number, data.shampoo, ':m', linewidth=2)
ax.plot(data.month_number, data.moisturizer, ':y', linewidth=2)

ax.set(xlabel='time (month)', ylabel='profit (€)',
       title='Total profit per month')
ax.legend(['Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap', 'Shampoo', 'Moisturizer'], loc='upper left')

plt.yticks([2000, 4000, 6000, 8000, 10000, 12000, 14000, 16000, 18000, 20000])

plt.xticks(data.month_number)
ax.grid()
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.

```python
fig, ax = plt.subplots()
plt.scatter(data.month_number, data.toothpaste, alpha=0.5)
ax.set(xlabel='time (month)', ylabel='profit (€)',
       title='Total profit per month')
ax.legend(['Toothpaste'], loc='upper left')

plt.yticks([4000, 6000, 8000, 10000])

plt.xticks(data.month_number)
ax.grid()
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
fig, ax = plt.subplots()

plt.bar([a-0.25 for a in data.month_number], data.facecream, width= 0.25, label = 'Face Cream sales data', align='edge')
plt.bar([a+0.25 for a in data.month_number], data.facewash, width= -0.25, label = 'Face Wash sales data', align='edge')
ax.set(xlabel='time (month)', ylabel='profit (€)',
       title='Sales per month and unit')

plt.legend(['Facecream', 'Facewash'], loc='upper left')

plt.xticks(data.month_number)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
```
